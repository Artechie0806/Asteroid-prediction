<hr><h2>roc_aoc</h2>

```python
from sklearn.metrics import roc_auc_score,roc_curve,auc
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_test_numeric = le.fit_transform(y_test)
y_prob = rf.predict_proba(x_test)[:, 1]  
roc_auc = roc_auc_score(y_test, y_prob)
print(f"ROC-AUC: {roc_auc}")
fpr, tpr, thresholds = roc_curve(y_test_numeric, y_prob)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualizing distributions for all features
for feature in ['moid_ld', 'H', 'diameter', 'q', 'i']:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='pha', y=feature, data=df_selective)
    plt.title(f'{feature} distribution by PHA class')
    plt.show()
```

```python

diameter_threshold = df_selective['diameter'].quantile([0.01, 0.99])
H_threshold = df_selective['H'].quantile([0.01, 0.99])
moid_ld_threshold = df_selective['moid_ld'].quantile([0.01, 0.99])

edge_cases = df_selective[
    (df_selective['diameter'] < diameter_threshold.iloc[0]) | 
    (df_selective['diameter'] > diameter_threshold.iloc[1]) |
    (df_selective['H'] < H_threshold.iloc[0]) | 
    (df_selective['H'] > H_threshold.iloc[1]) |
    (df_selective['moid_ld'] < moid_ld_threshold.iloc[0]) | 
    (df_selective['moid_ld'] > moid_ld_threshold.iloc[1])
]

print(edge_cases[['moid_ld', 'H', 'diameter', 'q', 'i', 'pha']])
```

```python

X_edge_cases = edge_cases[['moid_ld', 'H', 'diameter', 'q', 'i']]
y_edge_cases = edge_cases['pha']
y_pred_edge_cases = rf.predict(X_edge_cases)



print("Classification Report for Edge Cases:")
print(classification_report(y_edge_cases, y_pred_edge_cases))

cm_edge_cases = confusion_matrix(y_edge_cases, y_pred_edge_cases)
print("Confusion Matrix for Edge Cases:")
print(cm_edge_cases)
```
